```python
# Recreate all 6 visualizations and display them directly in the notebook (no file saving)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Arrests by offense, age, and race  Office of Juvenile Justice and Delinquency Prevention.csv")

# Convert numeric race columns
race_columns = ['White', 'Black', 'American Indian', 'Asian']
df[race_columns] = df[race_columns].replace(',', '', regex=True).apply(pd.to_numeric, errors='coerce')

# Filter datasets
aggravated = df[df['Offenses'].str.contains("Aggravated assault", case=False, na=False)]
simple_assault = df[df['Offenses'].str.contains("Simple assault", case=False, na=False)]
drug_abuse = df[df['Offenses'].str.contains("Drug abuse", case=False, na=False)]
murder = df[df['Offenses'].str.contains("Murder", case=False, na=False)]

# 1. Histogram of Aggravated Assault
plt.figure(figsize=(8, 6))
aggr_data = aggravated[race_columns].sum()
sns.barplot(x=aggr_data.index, y=aggr_data.values, palette="magma")
plt.title("Histogram of Aggravated Assault by Race")
plt.ylabel("Number of Arrests")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Stacked Bar Chart for Drug Abuse Violations
plt.figure(figsize=(10, 6))
drug_data = drug_abuse[race_columns]
drug_data.index = drug_abuse['Offenses']
drug_data.plot(kind='bar', stacked=True, colormap="tab20", figsize=(10, 6))
plt.title("Stacked Bar Chart: Drug Abuse Violations by Race")
plt.ylabel("Number of Arrests")
plt.xlabel("Offense Subtype")
plt.tight_layout()
plt.show()

# 3. Donut Chart for Simple Assault
plt.figure(figsize=(6, 6))
sa_data = simple_assault[race_columns].sum()
wedges, texts, autotexts = plt.pie(sa_data, labels=sa_data.index, autopct='%1.1f%%', startangle=90)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)
plt.title("Donut Chart: Simple Assault by Race")
plt.tight_layout()
plt.show()

# 4. Area Chart of All Offenses by Race
plt.figure(figsize=(12, 6))
area_data = df[race_columns].copy()
area_data['Offenses'] = df['Offenses']
area_data.dropna(inplace=True)
area_data.set_index('Offenses', inplace=True)
area_data.plot.area(figsize=(12, 6), colormap='Set2')
plt.title("Area Chart of All Offenses by Race")
plt.ylabel("Number of Arrests")
plt.xlabel("Offense Type")
plt.tight_layout()
plt.show()

# 5. Scatter Plot: Simple Assault vs Aggravated Assault
plt.figure(figsize=(8, 6))
x = simple_assault[race_columns].sum()
y = aggravated[race_columns].sum()
for race in race_columns:
    plt.scatter(x[race], y[race], label=race)
    plt.text(x[race], y[race], race)
plt.xlabel("Simple Assault Arrests")
plt.ylabel("Aggravated Assault Arrests")
plt.title("Scatter Plot: Simple vs Aggravated Assault by Race")
plt.legend()
plt.tight_layout()
plt.show()

# 6. Stacked Area Chart for Murder
plt.figure(figsize=(10, 6))
murder_data = murder[race_columns].copy()
murder_data['Offenses'] = murder['Offenses']
murder_data.set_index('Offenses', inplace=True)
murder_data.plot.area(stacked=True, colormap="Set3", figsize=(10, 6))
plt.title("Stacked Area Chart: Murder Offenses by Race")
plt.ylabel("Number of Arrests")
plt.xlabel("Offense Type")
plt.tight_layout()
plt.show()
```
